# Abstract

Presentiamo un approccio estremamente semplificato per la risoluzione del problema della predizione del punto di ancoraggio.
Definiamo una voxelgrid ${\bf V}$ come un tensore del terz'ordine <br>
<center>
       ${\bf V} := \sum_i^{N_x} \sum_j^{N_y} \sum_k^{N_z} V_{ijk} \ {\bf e^{(i)}} \otimes {\bf e^{(j)}} \otimes {\bf e^{(k)}} $
</center>
Chiamiamo $\mathscr V^*$ l'insieme di tutte le possibili voxelgrid.<br>
Chiamiamo $\mathscr V^{N_x,N_y,N_z} \subset \mathscr V^*$ l'insieme delle voxelgrid di dimensioni $N_x,N_y,N_z$

Il nostro dataset puó essere immaginato come segue
1. $\mathcal X = \{{\bf V}^{(1)},{\bf V}^{(2)},...,{\bf V}^{(N)}\} \subset \mathscr V^*$ una collezione di voxelgrid rappresentante le tasche
2. $\mathcal Y = \{{\bf C}^{(1)},{\bf C}^{(2)},...,{\bf C}^{(N)}\} \subset \mathscr V^*$ una collezione di voxelgrid, in cui ogni voxel contiene una probabilitá di attachment

Il dataset é costruito in modo tale che ${\bf V}^{(i)} \in \mathscr V^{a,b,c} \iff {\bf C}^{(i)} \in \mathscr V^{a,b,c} $


Definiamo l'operatore $\mathcal E_{n} :\mathscr V^* \rightarrow   \mathcal P ( \mathbb N \times  \mathscr V^{2n+1,2n+1,2n+1} ) $ che data in ingresso una qualsiasi voxelgrid ${\bf V}$ restituisce l'insieme di tutte le possibili sottogriglie di dimensione caratteristica $2n + 1$, ognuna <b> identificata da un numero intero </b>

Definiamo inoltre l'operatore $\mathscr C : \mathscr V^* \rightarrow \mathbb R$ che data una voxelgrid restituisce il valore del suo voxel centrale.
E.g $\mathscr V^{a,b,c} \ni {\bf V} \overset{\mathscr C}{\mapsto} V_{a/2,b/2,c/2}$

Possiamo costruire un nuovo dataset, adatto al nostro scopo, come segue
1. $X = \mathscr E_n(\mathcal X)$
2. $Y = \mathscr C ( \mathscr E_n (\mathcal Y)  )$

# Lo spazio $\mathscr V^{N,N,N}$ come spazio di Hilbert

Definiamo una nozione di prodotot scalare per lo spazio introdotto

$\langle \cdot, \cdot \rangle : \mathscr V^{N,N,N} \times \mathscr V^{N,N,N}
$
tale che
$ \langle {\bf V}, {\bf W} \rangle  = \sum_i^{N_x} \sum_j^{N_y} \sum_k^{N_z} V_{ijk} W_{ijk}
$


# Regressione logistica sullo spazio $\mathscr V^{N,N,N}$

Avendo definito un prodotto scalare possiamo risolvere il problema della Logistic Regression
<center>
    $\text{Find } {\bf W} \in \mathscr V^{N,N,N}, \beta \in \mathbb R \text{ minimizing } \sum_i \left( \sigma(\langle {\bf V}^{(i)}, {\bf W}\rangle) - C^{(i)} \right)^2$
</center>
