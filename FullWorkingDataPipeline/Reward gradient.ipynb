```python
import numpy as np
import jax.numpy as jnp
import jax

def corr(u,v):
    u_c = u - u.mean()
    v_c = v - v.mean()
    norma_u_c = jnp.linalg.norm(u_c)
    norma_v_c = jnp.linalg.norm(v_c)
    return (u_c @ v_c) / (norma_u_c * norma_v_c)
def r(W,X,Y):
    u = jax.vmap(lambda x: (x * W).sum() )(X)
    return corr(u,Y)
N = 10000
W = np.ones((3,3,3)); W[1,1,1] = -100.

X = np.random.normal(size = (N,3,3,3))
Y = jax.vmap(lambda x : x.sum())(X)

grad = jax.grad(r)
corretto = grad(W,X,Y)
```

```python
def handcrafted_grad(W,X,Y):
    X_bar = X.mean(axis = 0)
    Y_hat = jax.vmap(lambda x: (x * W).sum() )(X)
    A_1   = ((X - X_bar) * (Y - Y.mean())[:,None,None,None] ).sum(axis = 0)
    A     = A_1 / (jnp.linalg.norm(Y_hat - Y_hat.mean()) * jnp.linalg.norm(Y - Y.mean()))
    
    B_1   = ( (Y_hat - Y_hat.mean())[:,None,None,None]*(X - X_bar) ).sum(axis = 0)* ((Y_hat - Y_hat.mean()) @ (Y - Y.mean()))
    B_2   = jnp.linalg.norm(Y_hat - Y_hat.mean())**3 * jnp.linalg.norm(Y - Y.mean())
    B     = B_1/B_2
    return A - B
predetto = handcrafted_grad(W,X,Y)
```
